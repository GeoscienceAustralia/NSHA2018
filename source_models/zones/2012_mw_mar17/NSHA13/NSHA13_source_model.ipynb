```python
# Load Mark's source model file
#path = '.'
path = '/nas/gemd/ehp/georisk_earthquake/hazard/NSHM_18/PSHA_modelling/oq_inputs/NSHM'
source_model_file = os.path.join(path, 'source_model_NSHMandhotspots.xml')
parser = nrmlSourceModelParser(source_model_file)

# Parse the seismic sources and save them into a variable called "source_model"
source_model = parser.read_file("NSHM13 Source Model") # You need to supply a name for the source model
```

```python
# Plot the polygons
llon, ulon, llat, ulat = 105, 155, -45, -5,
map_config = {'min_lon': np.floor(llon), 'max_lon': np.ceil(ulon),
              'min_lat': np.floor(llat), 'max_lat': np.ceil(ulat), 'resolution':'i'}

# Adding the seismic sources
borders = ['r-', 'k-', 'b-', 'g-', 'y-', 'c-', 'm-', 'b-']
#orders = ['r', 'b', 'g']
#plt.gca().set_color_cycle(borders)
basemap1 = HMTKBaseMap(map_config, 'NSHM13 area source zones')
i=0
basemap1.fig.gca().set_color_cycle(borders)
for source in source_model:
    j = i%len(borders)
    if source.typology == 'Point':
        continue
    name = source.name[-4:]
#    print name
    basemap1._plot_area_source(source, border = borders[j], border_width=1.5, )#, alpha=0.5) 
    centre_lon = np.mean([max(source.geometry.lons), min(source.geometry.lons)])
    centre_lat = np.mean([max(source.geometry.lats), min(source.geometry.lats)])
    x,y = basemap1.m([centre_lon], [centre_lat])
#    print x,y,centre_lon, centre_lat
    basemap1.m.plot(x[0], y[0], 'bo', label=name)#, name)
 #   basemap1.fig.text(x[0], y[0], name)
    #print source.name
    i+=1

```
